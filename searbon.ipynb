```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = sns.load_dataset('tips')
data
```

```python
data.head()
```

```python
sns.set_theme()
```

```python
sns.relplot(data=data, x='total_bill', y='tip', col='time', hue='smoker', size='size')
```

```python
from sklearn.datasets import fetch_california_housing
data4 = fetch_california_housing()
```

```python
data4 = pd.DataFrame(data4.data, columns=data4.feature_names)
data4
```

```python
data4.head()
```

```python
sns.distplot(data4['MedInc'])
```

```python
correlation = data4.corr()
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(correlation, annot=True)
plt.show()
```
